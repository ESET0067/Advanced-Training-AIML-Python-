```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Set date as index
df = df.set_index('date')

print(df.head())
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df['humidity'], color='deepskyblue')
plt.title('Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df['meantemp'], color='red')
plt.title('meantemp Over Time')
plt.xlabel('Date')
plt.ylabel('meantemp')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df['wind_speed'], color='blue')
plt.title('wind_speed Over Time')
plt.xlabel('Date')
plt.ylabel('wind_speed')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df['meanpressure'], color='red')
plt.title('meanpressure Over Time')
plt.xlabel('Date')
plt.ylabel('meanpressure')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['meantemp', 'humidity', 'wind_speed', 'meanpressure']])
plt.show()
```
