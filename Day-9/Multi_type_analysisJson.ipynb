```python
# Convert stringified dictionaries/lists back to Python objects safely
def safe_eval(x):
    try:
        return ast.literal_eval(x)
    except:
        return np.nan

nested_cols = ['stock', 'dimensions', 'ratings']
for col in nested_cols:
    df[col] = df[col].apply(safe_eval)

# Expand nested JSON-like columns
for col in nested_cols:
    expanded = pd.json_normalize(df[col])
    expanded.columns = [f"{col}.{sub}" for sub in expanded.columns]
    df = pd.concat([df, expanded], axis=1)
```

```python
def clean_numeric(series):
    return pd.to_numeric(series.astype(str).str.replace(r'[^\d.]', '', regex=True),
                         errors='coerce')

num_cols = [
    'price', 'stock.quantity',
    'dimensions.weight', 'dimensions.width', 'dimensions.height', 'dimensions.depth',
    'ratings.averageRating', 'ratings.numberOfReviews'
]

for col in num_cols:
    if col in df.columns:
        df[col] = clean_numeric(df[col])
```
